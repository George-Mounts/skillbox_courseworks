## Part 0. Data preparation and overview

```python
import numpy as np
import pandas as pd
import os

# formal sh1t > actual work
```

```python
DATA_PATH = "./Data"
```

```python
df_dict = {}

for file_ in os.listdir(DATA_PATH):
    # read each csv table
    df_dict[file_] = pd.read_csv(os.path.join(DATA_PATH, file_))

# list of all tables
df_dict.keys()
```

### "Progresses" table

```python
# rename columns to dataset common standard
df_dict["progresses.csv"].columns = ["progress_id", "student_id", "course_id"]

# how much memory used by table
print(df_dict["progresses.csv"].memory_usage().sum() * 1e-6, "MB")

# sample
df_dict["progresses.csv"].sample(n=5, random_state=1)
```

### "Course Contents" table

```python
# how much memory used by table
print(df_dict["course_contents.csv"].memory_usage().sum() * 1e-6, "MB")

# sample
df_dict["course_contents.csv"].sample(n=5, random_state=1)
```

### "Students" table

```python
# rename columns to dataset common standard
df_dict["students.csv"].columns = ['id_', 'student_id', 'city', 'birthday']

# convert dates
df_dict["students.csv"]["birthday"] = pd.to_datetime(df_dict["students.csv"]["birthday"], errors="coerce")
print(df_dict["students.csv"].dtypes)

# how much memory used by table
print(df_dict["students.csv"].memory_usage().sum() * 1e-6, "MB")

# sample
df_dict["students.csv"].sample(n=5, random_state=1)
```

### "Progress Phases" table

```python
df_dict["progress_phases.csv"]["start_date"] = (
    pd.to_datetime(df_dict["progress_phases.csv"]["start_date"], errors="coerce", utc=False)
    .dt.tz_localize(None)
    )

df_dict["progress_phases.csv"]["finish_date"] = (
    pd.to_datetime(df_dict["progress_phases.csv"]["finish_date"], errors="coerce", utc=False)
    .dt.tz_localize(None)
    )

print(df_dict["progress_phases.csv"].dtypes)

# how much memory used by table
print(df_dict["progress_phases.csv"].memory_usage().sum() * 1e-6, "MB")

# sample
df_dict["progress_phases.csv"].sample(n=5, random_state=1)
```

### "Courses" table

```python
# rename columns to dataset common standard
df_dict["courses.csv"].columns = ["index", "course_id", "course_title", "course_field"]

# set index column as dataframe index
df_dict["courses.csv"].set_index("index", inplace=True)

# how much memory used by table
print(df_dict["courses.csv"].memory_usage().sum() * 1e-6, "MB")

# sample
df_dict["courses.csv"].sample(n=5, random_state=1)
```

### Merge tables

notes:

* "progresses" with "progress_phases" on "progress_id"
* dataset with "courses" on "course_id"
* dataset with "course_contents" on "module_number", "lesson_number" and "course_id"

```python
dataset = pd.merge(left=df_dict["progresses.csv"],
                   right=df_dict["progress_phases.csv"],
                   on="progress_id",
                   how="outer")

dataset = pd.merge(left=dataset,
                   right=df_dict["courses.csv"],
                   on="course_id",
                   how="outer")

dataset = pd.merge(left=dataset,
                   right=df_dict["course_contents.csv"],
                   on=["course_id", "lesson_number", "module_number"],
                   how="outer")

dset_students = df_dict["students.csv"].copy(deep=True)
```

### Copypasta to use in other parts of the coursework

```python
df_dict = {}

for file_ in os.listdir(DATA_PATH):
    df_dict[file_] = pd.read_csv(os.path.join(DATA_PATH, file_))
    
df_dict["progresses.csv"].columns = ["progress_id", "student_id", "course_id"]

df_dict["students.csv"].columns = ['id_', 'student_id', 'city', 'birthday']
df_dict["students.csv"]["birthday"] = pd.to_datetime(df_dict["students.csv"]["birthday"], errors="coerce")

df_dict["courses.csv"].columns = ["index", "course_id", "course_title", "course_field"]
df_dict["courses.csv"].set_index("index", inplace=True)

dataset = pd.merge(left=df_dict["progresses.csv"],
                   right=df_dict["progress_phases.csv"],
                   on="progress_id",
                   how="outer")

dataset = pd.merge(left=dataset,
                   right=df_dict["courses.csv"],
                   on="course_id",
                   how="outer")

dataset = pd.merge(left=dataset,
                   right=df_dict["course_contents.csv"],
                   on=["course_id", "lesson_number", "module_number"],
                   how="outer")

dataset["start_date"] = (
    pd.to_datetime(dataset["start_date"], errors="coerce", utc=False)
    .dt.tz_localize(None)
    )

dataset["finish_date"] = (
    pd.to_datetime(dataset["finish_date"], errors="coerce", utc=False)
    .dt.tz_localize(None)
    )
```
