для каждой записи:

    находится ли старт в данном периоде -> bool series

    просуммировать -> количество записей в данном периоде

`Прим.` не указано, является ли start_date датой получения студентом задания или же датой первой сдачи на проверку

формулировка задания подразумевает, что, скорее всего, 2 вариант верен

Проверка, есть ли "прогресс" в данном промежутке (period):

start_date < period.start_time < finish_date

OR

start_date < period.end_time < finish_date

Для каждого "прогресса":

```python
(
    (dataset["start_date"] <= per_range[0].start_time) &
    (dataset["finish_date"] >= per_range[0].start_time)
) | (
    (dataset["start_date"] <= per_range[0].end_time) &
    (dataset["finish_date"] >= per_range[0].end_time)
)
```

В каждом курсе надо

просуммировать прогрессы по каждому месяцу, на выходе - ряд

собрать полученные ряды в датафрейм, в котором колонки - курсы, индекс - периоды

```python
data_agg = pd.DataFrame()

for name, gr in dataset.groupby("course_title"):

    temp_df = pd.DataFrame()

    for per in per_range:
        temp_df[per] = (
            (
                (gr["start_date"] <= per.start_time) &
                (gr["finish_date"] >= per.start_time)
            ) | (
                (gr["start_date"] <= per.end_time) &
                (gr["finish_date"] >= per.end_time)
            )
        )

    data_agg[name] = temp_df.sum()
    
data_agg
```
