```python
### YOUR CODE HERE ###


# Working directory

DATA_PATH = "./Data"

import numpy as np
import pandas as pd
import os
```

```python

# make dataframes for each csv file and store it in dictionary
# key - filename, value - dataframe

df_dict = {}


# iterate over csv files in data directory

for file_ in os.listdir(DATA_PATH):
    df_dict[file_] = pd.read_csv(os.path.join(DATA_PATH, file_))
    
df_dict.keys()
```

```python
# quick peek at tables

df_dict["progresses.csv"].columns = ["progress_id", "student_id", "course_id"]

# control memory usage to prevent overflow
print(df_dict["progresses.csv"].memory_usage().sum() * 1e-6, "MB")

# 5 random records
df_dict["progresses.csv"].sample(n=5, random_state=1)
```

```python
print(df_dict["course_contents.csv"].memory_usage().sum() * 1e-6, "MB")

df_dict["course_contents.csv"].sample(n=5, random_state=1)
```

```python
print(df_dict["students.csv"].memory_usage().sum() * 1e-6, "MB")

df_dict["students.csv"].columns = ['id_', 'student_id', 'city', 'birthday']

df_dict["students.csv"].sample(n=5, random_state=1)
```

```python
print(df_dict["progress_phases.csv"].memory_usage().sum() * 1e-6, "MB")

df_dict["progress_phases.csv"].sample(n=5, random_state=1)
```

```python

# index column already in csv
# set it as dataframe index

df_dict["courses.csv"].columns = ["index", "course_id", "course_title", "course_field"]
df_dict["courses.csv"].set_index("index", inplace=True)

print(df_dict["courses.csv"].memory_usage().sum() * 1e-6, "MB")

df_dict["courses.csv"].sample(n=5, random_state=1)
```

```python
# overview
# looking for shared columns

summary_info = []

for df in df_dict:
    print(df,"\t\t", df_dict[df].columns.to_list())
    
    summary_info.append([df,
                df_dict[df].shape,
                df_dict[df].memory_usage().sum() * 1e-6
               ])
    
pd.DataFrame(summary_info,
             columns=["Dataframe", "Shape", "Size, MB"])
```

### Merge notes


progresses & progress_phases on "progress_id"

add courses on "course_id"

add course_contents on "module_number", "lesson_number" and "course_id"

```python

dataset = pd.merge(left=df_dict["progresses.csv"],
                   right=df_dict["progress_phases.csv"],
                   on="progress_id",
                   how="outer")


print(dataset.shape)
print(dataset.memory_usage().sum() * 1e-6, "MB")

# progresses + progress_phases
dataset.sample(n=5, random_state=1)
```

```python
dataset = pd.merge(left=dataset,
                   right=df_dict["courses.csv"],
                   on="course_id",
                   how="outer")


print(dataset.shape)
print(dataset.memory_usage().sum() * 1e-6, "MB")

# progresses + progress_phases + courses
dataset.sample(n=5, random_state=1)
```

```python

dataset = pd.merge(left=dataset,
                   right=df_dict["course_contents.csv"],
                   on=["course_id", "lesson_number", "module_number"],
                   how="outer")


print(dataset.shape)
print(dataset.memory_usage().sum() * 1e-6, "MB")

# progresses + progress_phases + courses + course_contents
dataset.sample(n=5, random_state=1)
```

```python
# c_dset means convenient dataset

c_dset = dset_courses[["course_title", "birthday"]]
```

```python
range_mask = (c_dset["birthday"] > np.datetime64("1955")) & (
    c_dset["birthday"] < np.datetime64("2008"))
```

### Максимальный возраст по курсам

```python
today = pd.Timestamp("2021")
list_df = []

for name, g in c_dset[range_mask].groupby("course_title"):
    ages = (today - g["birthday"]) / np.timedelta64(1, "Y")
    list_df.append([name, ages.max()])
    
pd.DataFrame(list_df, columns=["course_title", "age_max"])
```

### Минимальный возраст по курсам

```python
today = pd.Timestamp("2021")
list_df = []

for name, g in c_dset[range_mask].groupby("course_title"):
    ages = (today - g["birthday"]) / np.timedelta64(1, "Y")
    list_df.append([name, ages.min()])
    
pd.DataFrame(list_df, columns=["course_title", "age_min"])
```

### Средний возраст по курсам

```python
today = pd.Timestamp("2021")
list_df = []

for name, g in c_dset.groupby("course_title"):
    ages = (today - g["birthday"]) / np.timedelta64(1, "Y")
    list_df.append([name, ages.mean()])
    
pd.DataFrame(list_df, columns=["course_title", "age_mean"])
```

### Медианный возраст по курсам

```python
today = pd.Timestamp("2021")
list_df = []

for name, g in c_dset.groupby("course_title"):
    ages = (today - g["birthday"]) / np.timedelta64(1, "Y")
    list_df.append([name, ages.median()])
    
pd.DataFrame(list_df, columns=["course_title", "age_median"])
```

```python
dataset["student_id"].unique()
```

```python
start_dates = pd.DataFrame(
    dataset.groupby("student_id").apply(lambda x: x["start_date"].min())
)

start_dates.columns = ["first_hw"]
start_dates
```

```python
temp_dataset = dataset.drop_duplicates(subset="student_id")
temp_dataset = temp_dataset.merge(start_dates, on="student_id", how="inner")
print(temp_dataset.shape)
temp_dataset.head()
```

```python
per_range = pd.period_range("2016-03", "2019-07", freq="M")
per_range
```

```python
((per_range[0].start_time < temp_dataset["first_hw"]) & (temp_dataset["first_hw"] < per_range[0].end_time))
```

для каждой записи:

    находится ли старт в данном периоде -> bool series

    просуммировать -> количество записей в данном периоде
    
инициализировать новую колонку для каждого курса

пройти по каждому периоду, посчитать сумму

```python
data_agg = pd.DataFrame(index=per_range)

for name, gr_ in dataset.groupby("course_title"):
    
    # init new column
    data_agg[name] = 0
    
    for period in per_range:
        data_agg.loc[period, name] = (
            (
                period.start_time <= gr_["start_date"]) & (
                period.end_time >= gr_["start_date"]
            )
        ).sum()

data_agg
```

4. **(p)** Рассчитать количество прогрессов по выполнению домашних работ в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно) для каждого курса. Учитывать, что выполнение домашнего задания может перетекать из одного месяца в другой (такие дз надо включать в общее число прогрессов для всех месяцев, которые покрывает срок выполнения этих дз)


`Прим.` не указано, является ли start_date датой получения студентом задания или же датой первой сдачи на проверку

формулировка задания подразумевает, что, скорее всего, 2 вариант верен

Проверка, есть ли "прогресс" в данном промежутке (period):

start_date < period.start_time < finish_date

OR

start_date < period.end_time < finish_date

Для каждого "прогресса":

```python
(
    (dataset["start_date"] <= per_range[0].start_time) &
    (dataset["finish_date"] >= per_range[0].start_time)
) | (
    (dataset["start_date"] <= per_range[0].end_time) &
    (dataset["finish_date"] >= per_range[0].end_time)
)
```

В каждом курсе надо

просуммировать прогрессы по каждому месяцу, на выходе - ряд

собрать полученные ряды в датафрейм, в котором колонки - курсы, индекс - периоды

```python
data_agg = pd.DataFrame()

for name, gr in dataset.groupby("course_title"):

    temp_df = pd.DataFrame()

    for per in per_range:
        temp_df[per] = (
            (
                (gr["start_date"] <= per.start_time) &
                (gr["finish_date"] >= per.start_time)
            ) | (
                (gr["start_date"] <= per.end_time) &
                (gr["finish_date"] >= per.end_time)
            )
        )

    data_agg[name] = temp_df.sum()
    
data_agg
```
